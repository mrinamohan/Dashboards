## Import Libraries 

```python
import pandas as pd
import numpy as np
```

## Extract Data from CSV file

```python
marketing = pd.read_csv('marketing_data.csv')
marketing.head()
```

## Transform Data

```python
marketing.describe()
```

```python
#Removing spaces before/after the column names
marketing = marketing.rename(columns=lambda x: x.strip())
```

```python
# Look for null values
marketing.isnull().sum()
```

```python
# Percentage of missing values
percent_missing = marketing.isnull().sum() * 100 / len(marketing)
missing_value_df = pd.DataFrame({'column_name': marketing.columns,
 'percent_missing': percent_missing})
missing_value_df
```

```python
# Drop rows with missing values
marketing = marketing.dropna()
```

## Load transformed data to new csv for visualization

```python
marketing.to_csv('marketing_new_data.csv', index = False)
```
